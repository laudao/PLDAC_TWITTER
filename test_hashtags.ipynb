```python
twitter_db = connect_to_db()
```

```python
N = 100
hashtags_count = []

cursor = twitter_db.cursor()

query = ("SELECT h.hash, count(h.hash) \
          FROM hashs_new h, tweet_hash th \
          WHERE h.hash_id = th.hash_id \
          GROUP BY h.hash")

cursor.execute(query)

for h in cursor:
    hashtags_count.append(h)

cursor.close()
```
