```python
%load_ext autoreload
%autoreload 2
from utils import *
import pickle
```

```python
f = open("communities/partition", "rb")
partition = pickle.load(f)
f.close()
```

```python
all_users = list(partition.keys())
```

```python
nb_tweets_per_user = dict()
```

```python
twitter_db = connect_to_db()
```

```python
cursor = twitter_db.cursor()
query = "SELECT COUNT(*) FROM tweets WHERE user_id = " + str(all_users[0])
cursor.execute(query)
```

```python
cursor.fetchall()[0][0]
```

```python
for user in all_users:
    cursor = twitter_db.cursor()
    query = "SELECT COUNT(*) FROM tweets WHERE user_id = " + str(user)
    cursor.execute(query)
    nb_tweets_per_user[user] = cursor.fetchall()[0][0]
    cursor.close()
```
