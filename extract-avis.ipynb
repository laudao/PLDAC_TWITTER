```python
import os
from lxml import etree
import pandas as pd
%load_ext autoreload
%autoreload 2
from utils import *
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
path = 'decathlon/'

repos = os.listdir(path)
parser = etree.XMLParser(recover=True)
t=[]
for repo in repos:
    if os.path.isdir("decathlon/"+repo):
        files = os.listdir("decathlon/"+repo)
        for name in files:
            file=etree.parse("decathlon/"+repo+"/"+name, parser=parser)
            avis=file.xpath("./avis")
            for av in avis:
                d=dict()
                if len(av.xpath("./note/text()"))>0:
                    d["note"]=int(av.xpath("./note/text()")[0])
                    if d["note"]==3:
                        continue
                    #add label +1 -1 by note
                    if d["note"] >3:
                        d["label"]=+1
                    else:
                        d["label"]=-1
                else:
                    d["note"]=""
                if len(av.xpath("./commentaire/text()"))>0:
                    d["commentaire"]=av.xpath("./commentaire/text()")[0]
                else:
                    d["commentaire"]=""
                if len(av.xpath("./point_positif/text()")) > 0:
                    d["point_positif"]=av.xpath("./point_positif/text()")[0]
                else:
                    d["point_positif"]=""
                if len(av.xpath("./point_negatif/text()"))>0:
                    d["point_negatif"]=av.xpath("./point_negatif/text()")[0]
                else:
                    d["point_negatif"]=""

                t.append(d)
```

```python
data=pd.DataFrame.from_records(t)
```
